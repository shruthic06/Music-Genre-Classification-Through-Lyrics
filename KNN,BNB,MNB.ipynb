```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from sklearn.preprocessing import normalize
from nltk.stem import PorterStemmer
from sklearn.neighbors import KNeighborsClassifier

stemmer = PorterStemmer()

def pre_process():
    df = pd.read_csv("kaggle_processed.csv")
    df.dropna(subset=['Lyrics', 'Genre'], inplace=True)
    df = df[['Lyrics', 'Genre']] 

    labels = df['Genre'].tolist()
    text = df['Lyrics'].astype(str).str.lower().tolist()
    text = [" ".join([stemmer.stem(word) for word in sentence.split()]) for sentence in text]
    return text, labels

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

def train_and_evaluate(model_type):
    text, labels = pre_process()

    label_encoder = LabelEncoder()
    labels_encoded = label_encoder.fit_transform(labels)

    tfidf = TfidfVectorizer(stop_words='english', sublinear_tf=True)
    X_train, X_test, y_train, y_test = train_test_split(text, labels_encoded, test_size=0.2, random_state=42)

    X_train_vec = tfidf.fit_transform(X_train)
    X_test_vec = tfidf.transform(X_test)

    X_train_vec = normalize(X_train_vec)
    X_test_vec = normalize(X_test_vec)

    print(f"Training model: {model_type.upper()}")

    model = None
    if model_type == "mnb":
        model = MultinomialNB()
    elif model_type == "bnb":
        model = BernoulliNB()
    elif model_type == "lr":
        model = LogisticRegression(solver="liblinear", multi_class="ovr")
    elif model_type == "dt":
        model = DecisionTreeClassifier()
    elif model_type == "mlp":
        model = MLPClassifier(max_iter=300)
    elif model_type == "rf":
        model = RandomForestClassifier(n_estimators=100, max_features="sqrt")
    elif model_type == "svm":
        model = SVC(kernel="linear")
    elif model_type == "xgb":
        model = XGBClassifier(use_label_encoder=False, eval_metric="mlogloss")
    elif model_type == "knn":
        model = KNeighborsClassifier(n_neighbors=5) 
    else:
        print(f"Unsupported model type: {model_type}")
        return

    model.fit(X_train_vec, y_train)
    y_pred = model.predict(X_test_vec)

    print("\n Classification Report:")
    print(classification_report(y_test, y_pred, target_names=label_encoder.classes_, zero_division=0))

models_to_run = ["lr", "rf", "xgb", "svm", "mnb", "bnb", "dt", "mlp","knn"]

for model_name in models_to_run:
    train_and_evaluate(model_name)
```
